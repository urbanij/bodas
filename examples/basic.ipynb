````python
import bodas
````

````python
# import [SymPy](https://www.sympy.org) the Python library for symbolic mathematics
import sympy
````

````python
# define `s` as symbol
s = sympy.Symbol('s')
````

````python
# define a transfer function `H` 

H = (1+s/23)/(1+s/123)**2 
````

````python
H
````

````python
bodas.plot(H) # call the `plot` function defined in the bodas library
````

````python
bodas.plot(H, savefig='svg')   # 'svg' or 'png'
````
